```python
import os

print("🌟 FASE 1: Estrazione della Proteina d'Oro e del Ligando GPS...")

file_cif = "chai_vincitore.cif"

if not os.path.exists(file_cif):
    print(f"❌ ERRORE: Non trovo il file '{file_cif}'. Assicurati che sia nella stessa cartella.")
else:
    # 1. Convertiamo in PDB usando OpenBabel in locale
    print("🔄 Converto il file CIF in PDB...")
    os.system(f"obabel -icif {file_cif} -opdb -O complesso_temporaneo.pdb 2> /dev/null")

    # 2. Separazione delle coordinate
    print("✂️ Separo la proteina dal ligando...")
    with open("complesso_temporaneo.pdb", "r") as f:
        linee = f.readlines()

    proteina = [l for l in linee if l.startswith("ATOM")]
    ligando_gps = [l for l in linee if l.startswith("HETATM")]

    # 3. Salvataggio dei file operativi
    with open("proteina_gold.pdb", "w") as f:
        f.writelines(proteina)

    with open("ligando_gps.pdb", "w") as f:
        f.writelines(ligando_gps)
        
    os.remove("complesso_temporaneo.pdb")

    print("✅ Operazione completata con successo!")
    print("   - Creato: proteina_gold.pdb (La tua tasca perfetta)")
    print("   - Creato: ligando_gps.pdb (Il tuo mirino per Gnina)")
```
